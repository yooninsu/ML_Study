### 01. 지도 표현하기
#### 1) 지도 객체 생성
지도의 중심이 되는 위도아 경도를 설정

```python
# zoom_start: 배율 1~22
map_osm = folium.Map(location=[37.5004444, 127.0313861], zoom_start=17)
map_osm
```

#### 2) 지도객체에 마커 추가
##### 일반 마커
아이콘 색상값 종류
>'lightgreen', 'darkgreen', 'darkblue', 'cadetblue', 'orange', 'lightred', 'darkred', 'green', 'blue', 'black'
>'lightblue', 'white', 'lightgray', 'red', 'pink', 'beige', 'gray', 'purple', 'darkpurple'

```python
# 새로운 지도 객체 생성
map_osm1 = folium.Map(location = [43.663508952276075, -79.39595255866932], zoom_start = 17)

# 마커 객체 생성
# -> 마커가 표시될 위치 [위도, 경도], 클리시 표시될 메시지, 아이콘 색상
marker1 = folium.Marker([43.663508952276075, -79.39595255866932],
                                        popup="University of Toronto",
                                        icon = folium.Icon(color='beige')
)  
marker1.add_to(map_osm1)
map_osm1
```

#### 사용자 지정 아이콘, HTML 팝업

```python
# 새로운 지도 객체 생성
map_osm2 = folium.Map(location=[43.663508952276075, -79.39595255866932], zoom_start= 17)
 
# HTML을 사용한 팝업
popup_html = folium.Popup("<div style='white-space: nowrap'><h3>University of Toronto St. George Campus</h3><img src='https://lh5.googleusercontent.com/p/AF1QipO_zdzkuu9eD1RhOdgmVMRwaynsvxys8BSvuhyc=w203-h142-k-no'  width = '30%'><br/><p>tel: <a href='tel:+1 416-978-2011'>+1 416-978-2011</a></p><p>27 King's College Cir, Toronto, ON M5S 1A1, Canada</p></div>", parse_html=False)


# 사용자 지정 아이콘 이미지 사용
# --> 온라인 상의 URL, 내 컴퓨터 상대, 절대경로 모두 가능함.
icon_img =  folium.features.CustomIcon("https://data.hossam.kr/D04/ho.png", icon_size= (50,50))

# 마커 객체 생성
marker2 = folium.Marker([43.663508952276075, -79.39595255866932],
                                            popup=popup_html,
                                            icon=icon_img)
marker2.add_to(map_osm2) # 마커 객체를 지도에 추가함

# 웹 페이지 파일이 저장될 폴더 생성
if not path.exists('output'):
    mkdir('output')

map_osm2.save('output/map_osm2.html') #파일이 저장될 위치

map_osm2
```
