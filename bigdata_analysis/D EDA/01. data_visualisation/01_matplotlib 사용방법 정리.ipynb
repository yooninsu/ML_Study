#### Matplotlib 기본 사용 방법

#### 01.  Matplotlib 개요

Matplotlib는 파이썬에서 자료를 그래프로 시각화하는 패키지이다. <br>
Matplotlib에 포함된 pyplot이라는 객체를 통해 다양한 시각화 기능을 제공

#### 표현 가능한 그래프 종류
    - Line plot
    - Bar chart
    - Box plot
    - Pie chart
    - Histogram
    - Scatter Plot


```python
from matplotlib import pyplot as plt
```

#### 2. 기본 코드 형식

##### 그래프 시작을 선언
plt.figure()

#.... 그래프에 표현할 내용을 구현

##### (필요하다면) 그래프 이미지 저장하기
plt.savefig('저장할 png이미지의 경로', [dpi= 수치값])

### 02. 라인 플롯

가장 간단한 플롯은 선을 그리는 라인 플롯(Line plot)이다.<br>
라인 플롯은 데이터가 시간, 순서 등에 따라 어떻게 변화하는지 보여주기 위해 사용한다.<br>
명령은 pyplot객체의 `plot()` 매서드를 사용한다.

<br>

#### 기본 그리기
`plot()` 매서드에 전달되는 리스트나 numpy 배열의 인덱스가 x축이 되고 값이 y축이 된다.

```python
plt.figure()
plt.title('My Plot')
plt.plot([1,3,5,6,7,8,9])
plt.xlabel("X")
plt.ylabel("Y", rotation = 0) # To show Y horizontally
plt.show()
plt.close()
```

#### 2. X축 별도 지정
`plot()` 매서드에 전달되는 첫번째 리스트가 x축, 두번째 리스트가 y축을 담당한다.

```python
plt.figure()
plt.title('x ticks')
plt.plot([10,30,50,70], [1,2,3,4])
plt.xlabel("X")
plt.ylabel("Y", rotation = 0) # To show Y horizontally
plt.show()
plt.close()
```

#### 3. 한글 사용하기
한글을 사용하려면 pyplot객체에 한글 폰트를 설정해야 한다. <br>
당연히 해당 폰트는 컴퓨터에 깔려 있어야 한다.


```python
import os
import sys
from matplotlib import font_manager
```

```python
# 시스템 글꼴 목록을 리스트로 가져오기
font_list = font_manager.findSystemFonts()

# 이름순으로 정렬함
font_list.sort()

#font_list

# 각 폰트 파일의 경로에 대한 반복
for font_path in font_list:
    # 폰트 파일의 경로를 사용하여 폰트 속성 가져오기
    fp = font_manager.FontProperties(fname=font_path)
    #print(fp)

    try:
        font_name = fp.get_name()
        print(font_name)
    except:
        continue
```

#### 2. 글꼴을 적용하기 위한 그래프 전역 설정
이 설정은 같은 내용이 재설정되기 전까지 모든 그래프에 고옹 적용된다.

```python
# 폰트 설정
plt.rcParams['font.family'] = "Malgun Gothic"  # 윈도우 전용

# 글자 크기
plt.rcParams["font.size"] = 12

# 그래프 크기 설정(inch)
plt.rcParams["figure.figsize"] = (10, 5)

# 유니코드에서 음수 부호설정
plt.rcParams['axes.unicode_minus'] = False
```

#### 3. 한글 테스트

```python
plt.figure()
plt.title('한글 제목')
plt.xlabel('엑스축 라벨')
plt.plot('와이축 라벨')
plt.plot([-20, -10, 0, 10, 20])
plt.show()
plt.close()

```

#### 4. 각 영역별로 개별 폰트 지정

```python
# windows font
font1 = {'family': 'NanumGothic', 'size': 20, 'color': '#2aa2c9', 'weight': 'bold'}
font2 = {'family': 'Batang', 'size': 10, 'color': 'darkcyan'}
font3 = {'family': 'Batang', 'size': 10, 'color':  'darkblue', 'weight': 'light'}

plt.figure()
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
plt.close()
```

#### 04. 스타일 지정

풀롯 명령어는 보는 사람이 그림을 더 알아보기 쉽게 하기 위해 다양한 스타일을 지원한다. `plot` 명령어에서는 다음과 같이 추가 문자열 인수를 사용하여 스타일을 지원한다.

1. 선스타일

스타일 문자열	의미
-	실선(solid)
--	대시선(dashed)
:	점선(dotted)
-.	대시-점선(dash-dit)

```python
plt.figure()
plt.title("스타일 지정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 
         linestyle=':')
plt.grid()  # 그리드 표시
plt.show()
plt.close()
```

#### 2. 마커

데이터 위치를 나타내는 기호를 마커라고 한다.

```python
plt.figure()
plt.title("스타일 지정")
plt.plot([10,20,30,40], [1,4,9,15], linestyle= ':', marker="H")
plt.grid()
plt.show()
plt.close()
```

#### 3. 기타 스타일


```python
plt.figure()
plt.title("스타일 지정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 
         color="slateblue", linestyle='-', linewidth=5,
         marker='D', markersize=20, markerfacecolor="orange",
         markeredgecolor="violet", markeredgewidth=2)
plt.grid()  # 그리드 표시
plt.show()
plt.close()
```

### 05. 축 설정
#### 1. 범위 지정
x 축의 범위(`xlim`)와 y 축의 범위 (`ylim`)을 설정

```python
plt.figure()
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlim(10, 40)     # x축의 범위
plt.ylim(0, 20)   # y축의 범위
plt.show()
plt.close()
```

#### 2. 각 축의 표시 내용 설정

```python
y_values= list(range(70,101,5))
y_values
```

```python
y_ticks = []

for i in y_values:
    y_ticks.append("%d점" % i)

y_ticks
```

```python
plt.figure()
plt.title("학년별 평균 점수 변화")
plt.plot([1, 2, 3, 4], [98, 82, 73, 92])

# x축에 표시될 내용 설정 -> 좌표값 리스트와 표시 내용 리스트
plt.xticks([1, 2, 3, 4], ['1학년', '2학년', '3학년', '4학년'])

# y축에 표시될 내용 설정 -> 좌표값 리스트와 표시 내용 리스트
plt.yticks(y_values, y_ticks)

plt.grid()

plt.show()
plt.close()
```

### 06. 겹쳐 그리기
pyplot은 `figure()` 매서드를 호출한 이후 부터 `show()`, `close()` 매서드를 호출하기 전까지의 코드들이 셀로판지 위에 겹쳐지듯 그래픽을 표현한다.

```python
plt.figure()
plt.grid()
plt.title('복수의 plot 명령을 한 그림에서 표현')

plt.plot(range(0,11))
plt.plot([1,2,4,8,16,32,64,128,256,512])

plt.show()
plt.close()
```

### 07. 범례(legend)

여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지를 보여주기 위해 `legend` 명령으로 범례(legend)를 추가할 수 있다. 범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면 `loc` 인수를 사용한다. 인수에는 문자열 혹은 숫자가 들어가며 가능한 코드는 다음과 같다.


loc 문자열	숫자
best	0
upper right	1
upper left	2
lower left	3
lower right	4
right	5
center left	6
center right	7
lower center	8
upper center	9
center	10

```python
plt.figure()
plt.grid()
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([0, 10, 20, 30, 40, 50, 60, 70], label='1차 방정식')
plt.plot([0, 2, 4, 8, 16, 32, 64, 128], label='2차 방정식')
plt.legend(loc=2 )
plt.show()
plt.close()
```
