## 03. 공분산

공분산이란
>공분산은 두 개의 확률 변수의 선형관계를 나타내는 값이다. 한 확률 변수의 증감에 따른 다른 확률 변수의 증감의 경향에 대한 측도이다. 쉽게 말해 분산이라는 개념을 확장하여 두 개의 확률 변수의 흩어진 정도를 공분산이라고 하는 것이다.

공분산은 상관관계의 상승 혹은 하강하는 경향을 이해할 수는 있으나 2개 변수의 측정 단위의 크기에 따라 값이 달라지므로 절대적 정도를 파악하기에는 한계가 있다.

결과값이 양수면 정비례, 음수면 반비례

하지만 단위가 다른 데이터들 사이의 비교는 의미가 없기 때문에 두데이터의 분포가 동일하다고 해도 그에 대한 표준화가 필요하다. 즉 공분산을 표준화한 값이 피어슨 상관계수이다.

```python
np.cov(df['언어영역'], df['수리영역'])[0, 1]
```

## 04. 상관분석
### numpy를 통한 상관계수 구하기


$r = \frac{S_{xy}}{S_x \times S_y} = \frac{\sum_{i=0}^{n}[\left(x_i-\bar{x}\right)\left(y_i-\bar{y}\right)]}{\sum_{i=0}^{n}{(x_i-\bar{x})}  \sum_{i=0}^{n}{(y_i-\bar{y})}}$


```python
np.corrcoef(df['언어영역'], df['수리영역'])[0, 1]
```

#### 피어슨 상관계수와 p-value 구하기


```python
r, p = stats.pearsonr(df['언어영역'], df['수리영역'])

"상관계수: %f, p-value: %f" % (r, p)
```

#### 데이터프레임 객체를 통한 상관계수 구하기
numpy를 통해 구한 결과와 동일하다.

이와 같이 표시되는 형태를 상관행렬이라고 한다

```python
corr = df.corr(method='pearson')
corr
```

상관분석 결과 시각화

```python
plt.rcParams["figure.figsize"] = (5,3)

sb.heatmap(corr,
           annot = True, # 실제 화면에 나타낼 것인가
           cmap = 'Greens', # 색상
           vmin = -1, vmax= 1, # 컬러차트 영역 -1 ~ +1
           linewidths= .5 #구분선
)

plt.show()
plt.close()
```

#### 예제 (2) - mtcars 데이터를 통한 실린더수와 연비의 상관관계 분석
R 패키지인 ggplot2에 내장된 샘플 데이터로 자동차 32종의 11개 속성에 대한 정보를 담고 있는 데이터
| 필드 | 설명 |
|---|---|
| mpg | 연료 효율성 (마일 당 갤런) |
| cyl | 실린더 수 |
| disp | 배기량 (cubic inches) |
| hp | 마력 (horsepower) |
| drat | 후륜축 비율 (rear axle ratio) |
| wt | 차량 무게 (1000 파운드) |
| qsec | 1/4 마일 달성 시간 (초) |
| vs | V/S (V 엔진 형태인지, S는 직렬 엔진 형태인지를 나타냄) |
| am | 변속기 형태 (0: 자동, 1: 수동) |
| gear | 전진 기어 수 (전송 기어) |
| carb | 기화기 개수 (carburetors) |

```python
df2 = read_excel("https://data.hossam.kr/E03/mtcars.xlsx", index_col='name')
df2
```

#### 실린더 수에 따른 연비에 대한 산점도 그래프

```python
plt.rcParams["figure.figsize"] =(10, 5)

sb.regplot(data =df2, x = 'cyl', y = 'mpg')
plt.grid()
plt.show()
plt.close()
```

실린더 수가 증가함에 따라 연비가 낮아지는 경향이 있는것으로 보임



#### 상관분석

```python
plt.rcParams["figure.figsize"] = (10,5)

corr = df.filter(['cyl','mpg']).corr(method = 'pearson')
sb.heatmap(corr,
            annot = True,
            cmap = 'Blues',
            vmin= -1, vmax =1,
            linewidths = 0.5)
plt.show()
plt.close()
```

분석결과 상관계수가 -0.85로 나타나 실린더수와 연비는 강한 음의 상관관계가 있는 것으로 나타남



#### 분석결과 검정

```python
stats.pearsonr(df['cyl'], df['mpg'])
```

##### 결과보고
32종의 자동차를 대상으로 실린더수에 따른 연비를 조사한 결과 실린더 수와 연비는 강한 음의 상관관계를 갖는 것으로 타나났다. p-value가 유의수준 0.05보다 작으므로 이는 통계적으로 유의하다.
