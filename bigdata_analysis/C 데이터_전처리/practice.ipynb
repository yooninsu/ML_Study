#### 문제3
다음은 한 도시의 20개 중, 고교를 대상으로 조사한 흡연율 조사이다. 도시와 농촌별에따른 남여 평균 흡연율을 확인할 수 있는 데이터 피벗테이블로 구성하시오.

```python
q3 = read_excel("https://data.hossam.kr/test/q3.xlsx")
q3
```

```python
pdf3 = pivot_table(q3,
                   index = '지역',
                   columns='구성',
                   values= '흡연율',
                   aggfunc='mean')
pdf3
```
