```python
from pandas import DataFrame
```

```python
gradeDic = {
    "학년" : [1,2,1,3,4],
    "성별" : ['남자', '여자', '남자', '여자', '남자'],
    '국어' : [98, 88, 92, 63, 23],
    "영어" : [77, 90, 70, 60, 50],
    "수학" : [88, 62, 83, 31, 90],
    "과학" : [64, 72, 79, 70, 88]
 }

# -> 딕셔너리의 key는 DataFramedml column 이름이 된다.
df = DataFrame(gradeDic)
df
```

```python
# 인덱스 지정하기
df = DataFrame(gradeDic, index = ['철수','영희', '민철', '수현', '호영'])
df
```

```python
from pandas import read_csv
df = read_csv("https://data.hossam.kr/grade.csv", encoding = 'euc-kr')
df
```

```python
# 원본에는 변화가 없으며 결과가 적용된 결과가 새로운 데이터프레임으로 리턴된다.
df2 = df.set_index('이름')
df2
```

##### 인덱스 설정을 원본에 반영하기
###### 데이터프레임 메서드 중에서 `inplace = True`를 지원하는 경우 리턴값 없이 원본에 즉시 반영된다.

```python
df.set_index('이름', inplace=True)
df
```

##### 데이터 로딩시 인데스 열을 지정하기
###### 데이터파일을 가져오는 과정에서 인덱스로 사용할 열 이름을 미리 지정할 수 있다.
