```python
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as product
import time
import datetime as dict
import os
from urllib.parse import quote
```

```python
text = "제로 콜라"
encoded_text = quote(text)

print(encoded_text)
```

```python
# 크롬드라이버 자동 설치
chromedriver_autoinstaller.install()

driver = webdriver.Chrome()

driver.implicitly_wait(5)
```

```python
baseUrl = "https://browse.gmarket.co.kr/search?keyword=%s&k=30&p={}" % (encoded_text)
```

```python
result = []
for page_number in range(1, 11):
    url = baseUrl.format(page_number)
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    productList = soup.select(".section__inner-content-body-container > . box__information")
```

```python
searchList = soup.select(".box__information")
searchList
```

```python
result = []
for product in searchList:
    nameEl = product.select('.text__item')
    name = nameEl[0].text.strip()
    priceEl = product.select()

    item = {"이름": name}
    result.append(item)
```
